```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2019.csv")
df.head()
```

```python
df.isnull().any()
```

```python
df["노출(분)"]=df["노출(분)"].fillna(method='ffill')
df
```

```python
df.rename(columns={' 취급액 ':'취급액'}, inplace = True)
```

```python
df.columns
```

```python
# 무형 데이터 추출
dfformless = df[df['상품군'] == "무형"]
dfformless
```

```python
# df2에 무형데이터 제외한 데이터 삽입
idx_fl = df[df['상품군'] == "무형"].index
df2 = df.drop(idx_fl)
df2
```

```python
# 판매단가 및 취급액 데이터 앞뒤공백제거
df2['판매단가']  = df2['판매단가'].str.strip()
df2['취급액']  = df2['취급액'].str.strip()
df2
```

```python
# 판매단가 및 취급액 데이터 앞뒤공백, ","제거
df2['판매단가']  = df2['판매단가'].str.replace(",","")
df2['취급액']  = df2['취급액'].str.replace(",","")
df2
```

```python
df2.dtypes
```

```python
#판매단가 및 취급액 데이터 int로 변환
df2[['판매단가','취급액']] = df2[['판매단가','취급액']].apply(pd.to_numeric)
df2
```

```python
df2.dtypes
```

```python
from matplotlib import pyplot as plt
```

```python
#df2['취급액'].groupby(df['노출(분)']).mean().plot()
```

```python
#방송일시를 datetime object 로 변경
df2['방송일시'] = df2['방송일시'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d %H:%M'))
df2
```

```python
df2.dtypes
```

```python
df2['방송일시']
```

```python
df2.set_index(df2['방송일시'], inplace=True)
df2
```

```python
df2['판매수량'] = df2['취급액']/df2['판매단가']
df2
```

```python
# 시간별 판매수량의 합
df2.판매수량.resample('H').sum()
```

[ resample() 메소드의 시간 단위 구간 설정 ]

- 5분 단위 구간    : resample('5T')

- 10분 단위 구간  : resample('10T')

- 20분 단위 구간 : resample('20T')

- 1시간 단위 구간 : resample('1H')

- 1일 단위 구간    : resample('1D')

- 1주일 단위 구각 : resample('1W')

- 1달 단위 구간    : resample('1M')

- 1년 단위 구간    : resample('1Y')
